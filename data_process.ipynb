```python
#split validation dataset
classes = mvtec3d_classes()

data_set_path = '/home/liuj35/complete/mvtec3d/'
def load_dataset(img_path, dest_path=None):
    
    img_tot_paths = []
    gt_tot_paths = []
    tot_labels = []
    defect_types = os.listdir(img_path)

    for defect_type in defect_types:
        if defect_type == 'good':
            rgb_paths = glob.glob(os.path.join(img_path, defect_type, 'rgb') + "/*.png")
            tiff_paths = glob.glob(os.path.join(img_path, defect_type, 'xyz') + "/*.tiff")
            gt_paths = glob.glob(os.path.join(img_path, defect_type, 'gt') + "/*.png")
            rgb_paths.sort()
            tiff_paths.sort()
            gt_paths.sort()
            
            i = 0
            destination_path_rgb = os.path.join(dest_path, defect_type, 'rgb')
            if not os.path.exists(destination_path_rgb):
                os.makedirs(destination_path_rgb )
            destination_path_xyz = os.path.join(dest_path, defect_type, 'xyz')
            if not os.path.exists(destination_path_xyz):
                os.makedirs(destination_path_xyz)
            destination_path_gt = os.path.join(dest_path, defect_type, 'gt')
            if not os.path.exists(destination_path_gt):
                os.makedirs(destination_path_gt)
            while i < 5:
                shutil.move(rgb_paths[i], destination_path_rgb)
                shutil.move(tiff_paths[i], destination_path_xyz)
                shutil.move(gt_paths[i], destination_path_gt)
                i += 1
                
            
            
        else: 
            rgb_paths = glob.glob(os.path.join(img_path, defect_type, 'rgb') + "/*.png")
            tiff_paths = glob.glob(os.path.join(img_path, defect_type, 'xyz') + "/*.tiff")
            gt_paths = glob.glob(os.path.join(img_path, defect_type, 'gt') + "/*.png")
            rgb_paths.sort()
            tiff_paths.sort()
            gt_paths.sort()
            
            i = 0
            destination_path_rgb = os.path.join(dest_path, defect_type, 'rgb')
            if not os.path.exists(destination_path_rgb):
                os.makedirs(destination_path_rgb )
            destination_path_xyz = os.path.join(dest_path, defect_type, 'xyz')
            if not os.path.exists(destination_path_xyz):
                os.makedirs(destination_path_xyz)
            destination_path_gt = os.path.join(dest_path, defect_type, 'gt')
            if not os.path.exists(destination_path_gt):
                os.makedirs(destination_path_gt)
            
            while i < 5:
                shutil.move(rgb_paths[i], destination_path_rgb)
                shutil.move(tiff_paths[i], destination_path_xyz)
                shutil.move(gt_paths[i], destination_path_gt)
                i += 1

    assert len(img_tot_paths) == len(gt_tot_paths), "Something wrong with test and ground truth pair!"
    

    
    # shutil.move(source_path, destination_path
    # return img_tot_paths, gt_tot_paths, tot_labels
    return img_tot_paths


for cls in classes:
    dest_path  = os.path.join(data_set_path, cls, 'validation')
    img_path  = os.path.join(data_set_path, cls, 'test')
    load_dataset(img_path, dest_path)
```
